```python
import cv2, glob, numpy as np, matplotlib.pyplot as plt
```

```python
test, red, violet, black = glob.glob("colors/*")
```

```python
blackImgs, redImgs, violetImgs, testImgs = glob.glob(black+"/*.jpg"), glob.glob(red+"/*.jpg"), glob.glob(violet+"/*.jpg"), glob.glob(test+"/*.jpg")
```

```python
def hsvSum(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)/255.
    valSum = np.sum(img[:,:,2])
    return valSum
```

```python
blackSum, redSum, violetSum = [], [], []
```

```python
for x in blackImgs:
    blackSum.append(hsvSum(x))
blackSum = sum(blackSum)/len(blackSum)
```

```python
for x in redImgs:
    redSum.append(hsvSum(x))
redSum = sum(redSum)/len(redSum)
```

```python
for x in violetImgs:
    violetSum.append(hsvSum(x))
violetSum = sum(violetSum)/len(violetSum)
```
