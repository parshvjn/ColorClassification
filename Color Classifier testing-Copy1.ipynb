```python
import cv2, glob, numpy as np, matplotlib.pyplot as plt
```

```python
test, red, violet, black = glob.glob("colors/*")
```

```python
blackImgs, redImgs, violetImgs, testImgs = glob.glob(black+"/*.jpg"), glob.glob(red+"/*.jpg"), glob.glob(violet+"/*.jpg"), glob.glob(test+"/*.jpg")
```

```python
def hsvSum(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)/255.
    valSum = np.sum(img[:,:,2])
    return valSum
```

```python
def averageValue(blackImgs):
    blackSum = []
    for x in blackImgs:
        blackSum.append(hsvSum(x))
    return sum(blackSum)/len(blackSum)
```

```python
blackSum, redSum, violetSum = [], [], []
```

```python
baverage = averageValue(blackImgs)
```

```python
vaverage = averageValue(violetImgs)
```
